```python
import numpy as np
import pandas as pd 
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datos= pd.read_csv(r"C:\Users\juaaceve\Downloads\Advanced-Econometrics-MSc--master\Advanced-Econometrics-MSc--master\Bases de datos\GujaratiPorter71.txt", delimiter="\t")
datos.head()
```

```python
fig, ax = plt.subplots()
ax.scatter(datos["PGNP"], datos["CM"])
ax.set_xlabel("PNB pc")
ax.set_ylabel("Mortalidad Infantil")
ax.set_title("Relación entre variables", fontsize = 18, color = "red")
plt.show()
```

```python
modelo = sm.OLS(datos["CM"], datos["PGNP"])
resultados = modelo.fit()
print(resultados.summary())
```

```python
modelo2=sm.OLS(datos["CM"],sm.add_constant(datos["PGNP"]))
resultado2 = modelo2.fit()
print(resultado2.summary())
```

```python
valores_estimados = resultado2.predict()
valores_estimados
```

```python
fig, ax = plt.subplots()
ax.scatter(datos["PGNP"], datos["CM"], color = "blue")
ax.scatter(datos["PGNP"], valores_estimados, color = "red")
ax.set_xlabel("PNB pc")
ax.set_ylabel("Mortalidad Infantil")
ax.set_title("Relación entre variables", fontsize = 18, color = "red")
plt.show()
```

```python
datos["1/PGNP"] = 1/datos["PGNP"]
datos
```

```python
modelo3=sm.OLS(datos["CM"],sm.add_constant(datos["1/PGNP"]))
resultado3 = modelo3.fit()
print(resultado3.summary())
```

```python
valores_estimados2 = resultado3.predict()
valores_estimados2
```

```python
fig, ax = plt.subplots()
ax.scatter(datos["PGNP"], datos["CM"], color = "blue")
ax.scatter(datos["PGNP"], valores_estimados2, color = "red")
ax.set_xlabel("PNB pc")
ax.set_ylabel("Mortalidad Infantil")
ax.set_title("Relación entre variables", fontsize = 18, color = "red")
plt.show()
```
